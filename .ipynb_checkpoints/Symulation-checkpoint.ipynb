```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from random import randint
import random
```

```python
data = pd.read_csv('data.csv', sep=';')
data.columns = ['Position', 'Opening', 'Closing', '0', '0']
x = data['Position'].to_numpy()
y_open = data['Opening'].to_numpy()
y_close = data['Closing'].to_numpy()
data
```

```python
plt.figure(figsize=(15,4))
plt.subplot(121)
plt.plot(x,y_open,'g')
plt.title('Current usage (opening)')
plt.xlabel('Position')
plt.ylabel('Current [mA]')

plt.subplot(122)
plt.plot(x,y_close,'g')
plt.title('Current usage (closing)')
plt.xlabel('Position')
plt.ylabel('Current [mA]')
plt.show()
```

```python
# FUNCTIONS

def modify(y_open, start, stop, level):
    y_new = []
    iter = 0
    if start != 0:
        for j in range(start):
            y_new.append(y_open[j])
    for i in range(start, stop):
        if i < stop-((stop-start)/2):
            if level > 0:
                temp = y_open[i] + iter
                y_new.append(temp)
                iter += level
            else:
                temp = y_open[i] - iter
                y_new.append(temp)
                iter -= level
        else:
            if level > 0:
                temp = y_open[i] + iter
                y_new.append(temp)
                if y_open[i] < y_new[i]:
                    iter -= level/2
            else:
                temp = y_open[i] - iter
                y_new.append(temp)
                if y_open[i] > y_new[i]:
                    iter += level/2
    M = len(y_new)   
    y = np.concatenate((y_new, y_open[M:len(y_open)]), axis=None)
    
    return y


def random_modyfication(characteristics, nr_of_mods):
    y = characteristics
    iter = 0
    step = 0
    while iter < nr_of_mods:
        start = randint(0,205)
        stop = randint(0,205)
        if y[start] == y_open[start] and y[stop] == y_open[stop]:
            if start > stop:
                continue
            else:
                level = (random.random()*-1)+0.5
                y = modify(y,start,stop,level)
                iter += 1
                print('start=',start,'\t stop=',stop,'\t level=',level)
        else:
            continue
            iter += 1
            
    return y


def adapt(original_y, modified_y):
    count = 0
    new_tab = []
    for item in original_y:
        if item < modified_y[count]:
            new = ((item*0.99) + (modified_y[count]*0.01))
            new_tab.append(new)
        else:
            new = ((item*0.99) + (modified_y[count]*0.01))
            new_tab.append(new)
        count += 1
            
    return new_tab

save = False

def treshing(original_y, modified_y):
    global save
    i = 0
    for item in original_y:
        tresh = (abs(item - modified_y[i])*100)/item
        if tresh > 10:
            save = True
            break
        i += 1
    if save:
        return modified_y
    else:
        return original_y
    
```

```python
test = random_modyfication(y_open,5)
    
plt.figure(figsize=(30,10))
plt.plot(x,test, 'r', x,y_open,'g', x, adapt(y_open, test),'k--')
plt.show()
```

```python
# Zmiana caÅ‚oÅ›ci w czasie

Y = y_open
X = x
mod_Y = test
for i in range(200):  
    Y = adapt(Y,test)
    Z = treshing(y_open, Y)

plt.figure(figsize=(30,10))
plt.plot(x,test, 'r', x,y_open,'g', x, Z,'k--')
plt.show()
```
